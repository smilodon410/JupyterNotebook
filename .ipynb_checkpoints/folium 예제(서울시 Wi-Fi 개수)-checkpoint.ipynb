```python
import pandas as pd
import json
import folium
```

```python
geo_path = 'data/seoul.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
df = pd.read_csv('data/seoul_wifi.csv', encoding='CP949')
df
```

```python
df.columns
```

```python
df = df[['구명', '유형', '지역명', '설치위치(X좌표)', '설치위치(Y좌표)', '설치기관(회사)']]
df
```

```python
df.columns = (['구', '유형', '위치명', 'lat', 'lon', '설치회사'])
df
```

```python
count_gu = df['구'].value_counts()
count_gu
```

```python
map = folium.Map(location=[37.5102, 126.982],
                zoom_start=11)

map
```

```python
map = folium.Map(location=[37.5102, 126.982],
                zoom_start=11,
                 tiles='Stamen Toner')

map
```

```python
map = folium.Map(location=[37.5502, 126.982],
                zoom_start=10,
                #tiles='Stamen Toner'
                )

map.choropleth(geo_data=geo_str,
              data=count_gu,
               columns=['gu', 'counts'],
               fill_color='YlGnBu',
               key_on='feature.id'
              )

folium.LayerControl().add_to(map)

map
```
