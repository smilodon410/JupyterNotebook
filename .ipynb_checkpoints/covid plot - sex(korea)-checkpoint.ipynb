```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data/TimeGender.csv')
df
```

```python
df_sel = df[['date', 'sex', 'confirmed']]
df_sel
```

```python
df_male = df_sel['sex'] == 'male'
male = df_sel[df_male]
male
```

```python
df_female = df_sel['sex'] == 'female'
female = df_sel[df_female]
female
```

```python
plt.figure(figsize=(10, 8))
plt.xticks(range(0, 121, 30), rotation=45)
sns.lineplot(data=male, x='date', y='confirmed')
sns.lineplot(data=female, x='date', y='confirmed')
```
