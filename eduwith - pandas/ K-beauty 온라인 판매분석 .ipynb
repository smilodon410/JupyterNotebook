## 데이터 전처리
### 기간에서 연도 분리하기

```python
df["기간"]
```

* 강사님과 데이터셋의 차이가 있어서 개인적으로 수정하기

```python
df[df["기간"] == 'Unnamed: 29']
df[10800:]
```

```python
drop_num = range(10800, 11250)
a = []
for i in drop_num:
    a.append(i)
    
a
```

```python
df = df.drop(a)
df
```

```python
int("2019. 4/4".split('.')[0])
```

```python
# lambda 사용

df['연도'] = df['기간'].map(lambda x : int(x.split('.')[0]))
df.head()
```

```python
df['분기'] = df["기간"].map(lambda x : int(x.split()[1].split('/')[0]))
df.tail()
```

### 금액을 수치 데이터로 표연하기

```python
# 본 데이터 셋은 이미 변경된 상태

# df['백만원'] = df['백만원'].replace('-', pd.np.nan).astype(float)
# df.head()
```

### 필요없는 데이터 제거

```python
df = df[(df['국가(대륙)별'] != '합계') & (df['상품군별'] != '합계')].copy()
df.info()
```

```python
# 결측치 확인
df.isnull().sum()
```

## K-beauty 시각화
### 전체 상품군 판매액

```python
df_total = df[df["판매유형별"] == '계'].copy()
df_total.head()
```

```python
sns.lineplot(data=df_total, x='연도', y='백만원')
```

* legend를 그래프 밖에 그리기
    - plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    - stack overflow 참조

```python
# 상품군 별
sns.lineplot(data=df_total, x='연도', y='백만원', hue='상품군별')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# subplot 활용
# seaborn = replot

sns.relplot(data=df_total, x='연도', y='백만원', hue='상품군별', kind='line',
            col='상품군별', col_wrap=4)
```

```python
# 화장품 품목을 제외하기

df_sub = df_total[~df_total["상품군별"].isin(["화장품", "의류 및 패션관련 상품"])].copy()
```

```python
sns.relplot(data=df_sub, x="연도", y="백만원", hue="상품군별", col="상품군별", col_wrap=4, kind="line")
```

### 의류 및 패션관련 상품 온라인쇼핑 해외직접판매액

```python
df_fashion = df[(df["상품군별"] == '의류 및 패션 관련상품') & (df['판매유형별'] == '계')].copy()
df_fashion.head()
```

```python
plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x='기간', y='백만원', hue='국가(대륙)별')
```

```python
df_fashion2 = df[(df["상품군별"] == '의류 및 패션 관련상품') & (df['판매유형별'] != '계')].copy()

plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion2, x='기간', y='백만원', hue='판매유형별', ci=None)
```

### 데이터 집계하기

```python
# defalut : 대륙별 평균 금액

result = df_fashion.pivot_table(index='국가(대륙)별', columns='연도', values='백만원', aggfunc='sum')
result
```

### 연산결과를 시각적으로 보기(heatmap)

```python
plt.figure(figsize=(10, 6))
sns.heatmap(result, cmap='Blues', annot=True, fmt=".0f")
```
