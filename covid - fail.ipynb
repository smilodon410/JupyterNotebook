```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Global_confirmed_cases.csv')
df
```

```python
df.rename(columns={'Country/Region':'Date'}, inplace=True)

df
```

```python
sel = df.loc['3/30/20']
a = sel.sort_values(ascending=False)
sel_top = a.head(10)
sel_top
```

```python
fig, ax = plt.subplots(figsize=(4, 2.5), dpi=144)
colors = plt.cm.Dark2(range(6))
y = sel_top.index
width = sel_top.values
ax.barh(y=y, width=width, color=colors);
```

```python
def nice_axes(ax):
    ax.set_facecolor('.8')
    ax.tick_params(labelsize=8, length=0)
    ax.grid(True, axis='x', color='white')
    ax.set_axisbelow(True)
    [spine.set_visible(False) for spine in ax.spines.values()]
    
    nice_axes(ax)
    fig
```

```python
fig, ax_array = plt.subplots(nrows=1, ncols=3, figsize=(7, 2.5), dpi=144, tight_layout=True)
dates = ['3/29/20', '3/30/20', '3/31/20']
for ax, date in zip(ax_array, dates):
    s = df.loc[date].sort_values(ascending=False).head(10)
    ax.barh(y=s.index, width=s.values, color=colors)
    ax.set_title(date, fontsize='smaller')
```
