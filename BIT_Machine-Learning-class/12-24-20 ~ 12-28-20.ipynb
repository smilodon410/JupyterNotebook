### k-NN Classification

```python
from sklearn.model_selection import train_test_split
# k-NN 사용하기
# 데이터 불러오기
x, y = mglearn.datasets.make_forge()

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
# 2. qnsfbrl ghcnf
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)

# 3. 훈련하기
clf.fit(x_train, y_train)

# 4. 테스트 세트 예측
clf.predict(x_test)

# 5. 테스트 점수 확인
"{0:.2f}".format(clf.score(x_test, y_test))
```

* 86% 정확하게 분류했다

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

x_train, x_test, y_train, y_test = train_test_split(cancer.data,
                                                   cancer.target,
                                                   stratify=cancer.target, # 계층적 분할을 수핼할 데이터 지정
                                                   random_state=66)
```

```python
# 정확도를 기록하기 위한 리스트 생성
training_accuracy = []
test_accuracy = []

# 이웃의 개수를 설정(1 ~ 10개)
neighbor_settings = range(1, 11)

for n_neighbors in neighbor_settings:
    # 모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    
    # 훈련
    clf.fit(x_train, y_train)
    
    # 정확도 저장하기
    training_accuracy.append(clf.score(x_train, y_train))
    test_accuracy.append(clf.score(x_test, y_test))
```

```python
# 시각화
plt.plot(neighbor_settings, training_accuracy, label="training accuracy")
plt.plot(neighbor_settings, test_accuracy, label="test accuracy")

plt.ylabel("accuracy")
plt.xlabel("n_neighbors")
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor

x, y = mglearn.datasets.make_wave(n_samples=40)

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

reg = KNeighborsRegressor(n_neighbors=3)

reg.fit(x_train, y_train)

reg.predict(x_test)

"{0:.2f}".format(reg.score(x_test, y_test))
```

$R^2$은 결정계수라고 합니다. 회귀에서의 스코링
* $y$ : target
* $\bar y$ : 훈련된 데이터의 평균

```python
import numpy as np
fig, axes = plt.subplots(1,4, figsize=(20,4))

#-3과 3 사이에 1000개의 데이터 만들기 -> 테스트 용도로 사용함
line = np.linspace(-3, 3, 1000).reshape(-1, 1)

for n_neighbors, ax in zip([1,3,5,9], axes):
    # 이웃의 개수를 1,3,9로 하여 예측하는 모델을 만듭니다
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(x_train, y_train)
    ax.plot(line, reg.predict(line)) # 테스트 용도로 만든 데이터를 예측하고 예측 결과를 선으로 표현합니다.
    ax.plot(x_train, y_train, '^', c=mglearn.cm2(0), markersize=8) # 훈련 데이터를 그래프에 표시합니다.
    ax.plot(x_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)   # 테스트 데이터를 그래프에 표시합니다.
    # 훈련 데이터의 점수와 테스트 데이터의 점수를 제목에 표현합니다.
    ax.set_title("{} neighbor's train score: {:.2f} / test score : {:.2f}".format(n_neighbors, reg.score(x_train, y_train), reg.score(x_test, y_test)))
    ax.set_xlabel("feature")
    ax.set_ylabel("target")
axes[0].legend(["model predicted", "train data/target", "test data/target"], loc="best")
```

### 가중치 페널티( $\alpha$ )
* $\alpha$란 가중치를 억제하는 힘 : $\alpha$가 커질 수록 $W$는 0에 가깝게 작아진다

    * Ridge 모델
        - $\alpha$가 커지면 $W$(가중치)가 0에 가깝게 작아지지만, 실제 0이 되지 않는다
    * Lasso 모델
        - $\alpha$가 커지면 $W$(가중치)가 0에 가까워 지고, 실제 0이 된다
        - 이러한 특성을 ***모델에 의한 특성***이라고 한다

```python
# 사용 데이터 셋 : 확장된 보스턴 주택가격 데이터 셋

# Ridge 확인하기
from sklearn.linear_model import Ridge

ridge = Ridge().fit(x_train, y_train)

print("훈련 세트 점수 : {:.2f}".format(ridge.score(x_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(ridge.score(x_test, y_test)))
```

```python
# alpha 값에 따른 Ridge 모델의 성능 확인

# alpha = 10 인 경우
ridge10 = Ridge(alpha=10).fit(x_train, y_train)
print("훈련 세트 점수 : {:.2f}".format(ridge10.score(x_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(ridge10.score(x_test, y_test)))
```

```python
# alpha = 0.5 인 경우
ridge01 = Ridge(alpha=0.1).fit(x_train, y_train)
print("훈련 세트 점수 : {:.2f}".format(ridge01.score(x_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(ridge01.score(x_test, y_test)))
```

```python
# alpha 값에 따른 각각 특성에 대한 가중치의 분포도 확인
# 선형 모델의 coef_ 변수에 가중치 배열 존재

plt.figure(figsize=(20, 10))

plt.plot(ridge10.coef_, '^', label="Ridge alpha 10")
plt.plot(ridge01.coef_, 's', label="Ridge alpha 0.1")
plt.plot(ridge.coef_, 'v', label="Ridge alpha 1")
plt.plot(lr.coef_, 'o', label="LinearRegression")

plt.xlabel("coef list")
plt.ylabel("coef size")
plt.hlines(0, 0, len(lr.coef_))
plt.ylim(-25, 25)
plt.legend()
plt.show()
```

```python
# Lasso
from sklearn.linear_model import Lasso

lasso = Lasso().fit(x_train, y_train)
print("훈련 세트 점수 : {:.2f}".format(lasso.score(x_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lasso.score(x_test, y_test)))
```

```python
# Lasso 는 모든 feature를 사용하지 않는다

print("Lasso 에서 사용한 특성의 개수 : {}".format(np.sum(lasso.coef_ !=0)))
```

```python
# Lasso 는 기본적으로 '경사하강법'으로 오차 수정
# max_iter = Lasso 가 수행할 졍사하강법의 최대 횟수

lasso001 = Lasso(alpha=0.01, max_iter=1000000).fit(x_train, y_train)
print("훈련 세트 점수 : {:.2f}".format(lasso001.score(x_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lasso001.score(x_test, y_test)))
print("Lasso001 에서 사용한 특성의 개수 : {}".format(np.sum(lasso001.coef_ !=0)))
```

## Linear Model Classification
* 이진 선형 분류기는 직선, 평면, 초평면을 활용해 클라스를 구분 지어준다
* 적절한 $W$와 $b$를 찾아서 그어주는 방법과 
    - 딥러닝에서 사용하는 방식 : 손실 함수( loss function 사용 ) - LogisticRegression, ...
* $W$와 $b$를 조절해 가며 규제를 사용하는 방법이 있다
    - SVM ( Support Vector Machine )

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

x, y = mglearn.datasets.make_forge()

fig, axes = plt.subplots(1, 2, figsize=(20, 10))

for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
    clf = model.fit(x, y)
    mglearn.plots.plot_2d_separator(clf, x, fill=False, eps=0.5, ax=ax, alpha=.7)
    mglearn.discrete_scatter(x[:, 0], x[:, 1], y, ax=ax)
    ax.set_title("{}".format(clf.__class__.__name__))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")

axes[0].legend()
plt.show()
```

* LinearSVC => c라는 규제 매개변수를 이용한다. 
    - $\alpha$와 비슷 : 가중치 조정할 수 있는 매개변수)
* LogisticRegression 
    - logitic( Sigmoid ) 함수를 활용 : loss가 최소화 되는 방향으로 가중치와 편향을 조절

```python
# svc 분류 그래프 확인
# 규제 매개변수 C를 활용한다 => alpha 의 반대 개념
# c가 증가하면 규제가 감소하면서 가중치가 증가
# c가 감소하면 규제가 증가하면서 가중치가 감소

mglearn.plots.plot_linear_svc_regularization()
```

* c ↓ => 모델이 단순(복잡도 하락) => 데이터의 패턴을 분석
* c ↑ => 모델이 복잡(복잡도 상승) => 데이터 포인트 각각을 신경

```python
def print_score(model, x, y):
    x_train, x_test = x
    y_train, y_test = y
    
    print("훈련 세트 점수 : {:.3f}".format(model.score(x_train, y_train)))
    print("테스트 세트 점수 : {:.3f}".format(model.score(x_test, y_test)))
```

```python
cancer = load_breast_cancer()

x_train, x_test, y_train, y_test = train_test_split(cancer.data,
                                                   cancer.target,
                                                   stratify=cancer.target,
                                                   random_state=42)

logreg = LogisticRegression(max_iter=10000).fit(x_train, y_train)

print_score(logreg, x=(x_train, x_test), y=(y_train, y_test))
```

```python
logreg100 = LogisticRegression(C=100, max_iter=10000).fit(x_train, y_train)
print_score(logreg100, x=(x_train, x_test), y=(y_train, y_test))
```

```python
logreg001 = LogisticRegression(C=0.01, max_iter=10000).fit(x_train, y_train)
print_score(logreg001, x=(x_train, x_test), y=(y_train, y_test))
```

### Linear Model Multiple Classification
* 이진 분류 : 0 아니면 1
* 다중 분류 : 0, 1, 2 ...

기본 머신러닝 모델은 binary classification만 지원하고, multiple classification 은 지원하지 않는다.

0, 1, 2 세 종류의 클래스가 있다고 가정하면, 0과 1을 비교하고, 1, 2를 비교하고, 2, 0을 비교해서 ***확률***이 가장 큰 클래스를 최종 결과물로 선정

One VS All( OVA ) 또는  One VS Rest (OVR)

이 때 사용되어지는 수식 = softmax 함수

다중 분류 함수도 이진 분류 선형함수와 같다.

$$
w[0] * x[0] + w[1] * x[1] + ... + w[p] * x[p] + b
$$

다중 분류가 되었을 때 각 클래스로 분류될 확률은 소프트맥스 함수에 의해 결정

$$
Pr(Y_i=c) = \frac{e^{W_c*X_i}}{\sum_{k=1}^K{e^{W_k*X_i}}}
$$


```python
from sklearn.datasets import make_blobs

x, y = make_blobs(random_state=42)

mglearn.discrete_scatter(x[:,0], x[:,1], y)
plt.xlabel("feature0")
plt.ylabel("feature1")
plt.legend(["class0", "class1", "class2"])
plt.show()
```
